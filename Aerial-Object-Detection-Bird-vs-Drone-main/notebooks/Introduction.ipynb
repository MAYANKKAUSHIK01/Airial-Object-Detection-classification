# **Project Summary -**

This project focuses on detecting and classifying aerial objects using deep learning models and deploying them through a Streamlit web application.

In modern surveillance systems, distinguishing between birds and drones is a major challenge.

Traditional systems often produce false alarms because birds and drones can appear visually similar in the sky.

With the increasing use of drones in areas like security, delivery, and monitoring, it becomes important to build an intelligent system that can automatically detect and classify these objects accurately

The main objectives of this project are:

•	To build a classification model that can identify whether an object is a bird or a drone 

•	To implement object detection using YOLO 

•	To create a real-time prediction system 

•	And to deploy everything using a Streamlit web interface

This makes the system interactive and easy to use.

# **Problem Statement**


**This project aims to develop a deep learning-based solution that can classify aerial images into two categories — Bird or Drone — and optionally perform object detection to locate and label these objects in real-world scenes.**

**The solution will help in security surveillance, wildlife protection, and airspace safety where accurate identification between drones and birds is critical. The project involves building a Custom CNN classification model, leveraging transfer learning, and optionally implementing YOLOv8 for real-time object detection. The final solution will be deployed using Streamlit for interactive use.**
